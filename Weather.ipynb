```python
import gradio as gr
import requests
import json

def get_weather(CITY):
    url = "https://yahoo-weather5.p.rapidapi.com/weather"
    querystring = {"location":CITY, "format": "json", "u": "c"}
    headers = {
        "X-RapidAPI-Key": "0a7b57a03emsh900e200e7213c95p1fff42jsn99287d9b2e90",
        "X-RapidAPI-Host": "yahoo-weather5.p.rapidapi.com"
    }
    response = requests.request("GET",url, headers=headers, params=querystring)
    response=response.json()
    #return response.text
    return str(response["current_observation"]["condition"]["temperature"])+" CELSIUS "
    

iface = gr.Interface(fn=get_weather,inputs="text",outputs="text",title="Yahoo Weather", description="Enter a city to get the weather.")
iface.launch()
```
